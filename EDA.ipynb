**IMPORTING LIBRARIES**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

**Reading the DataSet**

---



```python
my_data = pd.read_csv('rain_in_india.csv' )
my_data.head(7)
```

```python
my_data.isnull()
```

**LINE PLOT**

---



```python
sns.lineplot(data=my_data.head(6),x="Jan-Feb",y="SUBDIVISION")
plt.show()
```

**HEAT MAP-[NULL VALUES]**

---



```python
sns.heatmap(my_data.head(6).isnull())
```

**COUNT PLOT**

---



```python
sns.set_style('darkgrid')
sns.countplot(x="ANNUAL",hue="SUBDIVISION",data=my_data[25:28],palette='OrRd')
```

```python
sns.countplot(x="MAY",data=my_data[28:32],palette="BuGn")
```

**DISPLOT**

---



```python
sns.displot(my_data['ANNUAL'].dropna(),color='darkblue',bins=5)
```

**BAR PLOT**

---



```python
sns.barplot(data=my_data[28:34],x='Jun-Sep',y="SUBDIVISION")
```

**BOX-PLOT**

---



```python
plt.figure(figsize=(8,6))
sns.boxplot(x="Jan-Feb",y="Jun-Sep",data=my_data.head(15),palette='PiYG')
```

**SCATTER PLOT**

---



```python
plt.figure(figsize=(10,8))
sns.scatterplot(x="JAN",y="DEC",data=my_data,palette='Darkblue')
```

```python
sns.pairplot(my_data[0:51])
```

**CAT_PLOT**

---



```python
sns.catplot(data=my_data[0:10],x="AUG",y="SUBDIVISION")
```
